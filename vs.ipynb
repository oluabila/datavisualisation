```python
import urllib.request
import csv
import pandas as pd
file = "https://raw.githubusercontent.com/jldbc/coffee-quality-database/2e542fa4c87390771c55ecb3568d1e9518ce7b0e/data/arabica_data_cleaned.csv"
data = pd.read_csv(file, sep= ',')
data
```

```python
data.dtypes
```

```python
#data1 = data['Acidity','Aroma','Flavor','Sweetness','altitude_mean_meters']
#data2 = data['Acidity','Aroma']
df = data[['Acidity','Aroma','Flavor','Sweetness','altitude_mean_meters','Country.of.Origin','altitude_low_meters','altitude_high_meters','Total.Cup.Points']]
df
```

```python
print(df.isnull().sum())
```

```python
df = df.rename(columns={'Country.of.Origin': 'Country','Total.Cup.Points':'TotalCupPoints'})

df
```

```python
df = df.dropna(subset=['altitude_mean_meters','altitude_low_meters','altitude_high_meters','Country'])
```

```python
print(df.isnull().sum())
```

```python
df.isin([0]).sum()
```

```python
df_null = df [(df ['Aroma'] == 0) & (df ['Acidity'] == 0) & (df ['Flavor'] == 0) & (df ['Sweetness'] == 0)&(df ['Acidity'] == 0)]
df=df.drop(df_null.index, axis=0)

df.isin([0]).sum()
```
